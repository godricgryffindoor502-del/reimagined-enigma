```python
def quicksort(arr):
  if len(arr) <= 1:
    return arr
  pivot =arr[len(arr) // 2]
  left = [x for x in arr if x < pivot]
  middle=[x for x in arr if x == pivot]
  right = [x for x in arr if x > pivot]
  return quicksort(left) + middle + quicksort(right)
print(quicksort([1,5,2,5,3,90,2,44,7]))
```

```python
class Greeter:
  def __init__(self,name):
    self.name=name
  def greet(self,loud=False):
     if loud:
        print("HELLO, %s!" % self.name.upper())
     else:
        print("Hello, %s!" % self.name.capitalize())
g=Greeter('Dhruv')
g.greet()
g.greet(loud=True)

```

```python
def make_nd_array(shape,func=lambda:1):
  if len(shape)==1:
    return [func() for _ in range(shape[0])]
  return [make_nd_array(shape[1:],func) for _ in range(shape[0])]

```

```python
import imageio
import numpy as np
from PIL import Image
import requests
from io import BytesIO

# Load image from URL
url = "https://fastly.picsum.photos/id/237/536/354.jpg?hmac=i0yVXW1ORpyCZpQ-CknuyV-jbtU7_x9EBQVhvT5aRr0"
response = requests.get(url)
img = Image.open(BytesIO(response.content))
img=img.resize((200,200))
# Convert to numpy array
img = np.array(img)

print(img.dtype, img.shape)
```

```python
img_tinted = img * [0.8,0.8,0.8]
```

```python
img_tinted = np.clip(img_tinted, 0, 255)    # Ensure values in 0-255
img_tinted = img_tinted.astype(np.uint8)    # Convert back to uint8
```

```python
Image.fromarray(img_tinted).save("resized_image.jpg")
```
